```python
# 정규 방정식
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
X_b = np.c_[np.ones((100, 1)), X] # 모든 샘플에 x0 = 1을 추가한다.
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(theta_best)

# 절편 = 4.01159278, x계수 = 2.82698874
# y = 4 + 3x + 가우시안 잡음
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)  # (2*2) * (2*1)
y_predict
```

```python
import matplotlib.pyplot as plt
plt.plot(X_new, y_predict, 'r-')
plt.plot(X, y, 'b.')
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
print(lr.intercept_, lr.coef_)
print(lr.predict(X_new))
```

경사하강법 : 비용함수를 최소화하기 위해 반복해 파라미터 조정, 특성 수에 민감하지 않음

```python
# 배치경사하강법 : 매 스텝에서 훈련 데이터 전체 사용
eta = 0.1 # 학습률
n_iterations = 1000 # 스텝 수
m = 100 # 샘플 수

theta = np.random.randn(2, 1) # 무작위 초기화
for iteration in range(n_iterations) :
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - eta * gradients
```

```python
print(theta)  # 정규방정식과 같다
```

```python
# 확률적 경사하강법 : 매 스텝에서 한 개의 샘플을 무작위로 선택하고 그 하나의 샘플에 대한 그레이디언트 계산
n_epochs = 50
t0, t1 = 5, 50 # 학습 스케줄 하이퍼파라미터

def learning_schedule(t) : 
    return t0 / (t + t1)

theta = np.random.randn(2, 1) # 무작위 초기화

for epoch in range(n_epochs) :
    for i in range(m) : # 한 반복에서 m(샘플 수)번 되풀이
        random_index = np.random.randint(m)  # m개의 샘플 중 하나도 선택 못받을 수도, 중복되어 선택받을 수도(부트스트랩)
        xi = X_b[random_index : random_index+1]  # 한 개의 샘플
        yi = y[random_index : random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta)-yi)
        eta = learning_schedule(epoch * m + i) 
        theta = theta - eta * gradients
```

```python
# SGDRegressor : 사이킷런에서 SGD방식으로 선형 회귀 사용하려면 기본값으로 제곱 오차 비용함수를 최적화하는 클래스
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor(max_iter = 1000, tol = 1e-3, penalty = None, eta0=0.1)
# max_iter : 최대 에포크, tol : 손실이 tol만큼 줄어들 때까지 실행, penalty : 규제, eta0 : 학습률 
sgd.fit(X, y.ravel())  # ravel() : 1차원으로 변경
```

```python
# 2차 다항 새로운 특성 추가
# interaction_only = True : 거듭제곱이 포함한 항은 모두 제외
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 2, include_bias = False)
X_poly = poly_features.fit_transform(X) # 원래 특성 X와 이 특성의 제곱
print(X[0])
print(X_poly[0])
```

```python
lr = LinearRegression()
lr.fit(X_poly, y)
lr.intercept_, lr.coef_

# 원래 함수 y = 0.5x**2 + 1.0x + 2.0 + 가우시안잡음
# 예측된 함수 y_hat = 0.531x**2 + 0.964x + 1.947
```

```python
# 모델의 학습 곡선
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y) :
    train_input, val_input, train_target, val_target = train_test_split(X, y, test_size=0.2)
    #train_errors = []
    #val_errors=[]
    train_errors, val_errors = [], []
    for m in range(1, len(train_input) + 1) :
        model.fit(train_input[:m], train_target[:m])
        train_pred = model.predict(train_input[:m])
        val_pred = model.predict(val_input)
        train_errors.append(mean_squared_error(train_target[:m], train_pred))
        val_errors.append(mean_squared_error(val_target, val_pred))
    plt.plot(np.sqrt(train_errors), 'r-+', linewidth=2, label='훈련셋')
    plt.plot(np.sqrt(val_errors), 'b-', linewidth=3, label='검증셋')
```

```python
# 선형 회귀
lr = LinearRegression()
plot_learning_curves(lr, X, y)
plt.axis([0, 80, 0, 3])
plt.show()
# 과소적합 ; RMSE가 높아서
```

```python
# 10차 다항 회귀
from sklearn.pipeline import Pipeline
polynomial_regression = Pipeline([
    ('poly_features', PolynomialFeatures(degree=10, include_bias=False)),
    ('lin_reg', LinearRegression())
])
plot_learning_curves(polynomial_regression, X, y)
plt.axis([0, 80, 0, 3])
plt.show()

# 과대적합 ; 훈련데이터 오차가 선형 회귀보다 훨씬 낮다.
# 두 곡선 사이 공간 -> 훈련데이터에서의 모델 성능이 검증 데이터 보다 좋다.
# 과대적합 개선하는 방법은 검증 오차가 훈련오차에 근접할 때까지 더 많은 훈련 데이터를 추가하는 것
```

```python
# 조기 종료
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 2 + X + 0.5 * X**2 + np.random.randn(m, 1)

train_input, val_input, train_target, val_target = train_test_split(X[:50], y[:50].ravel(), test_size=0.5)

from copy import deepcopy
from sklearn.preprocessing import StandardScaler
from numpy import inf

poly_scaler = Pipeline([
    ('poly_features', PolynomialFeatures(degree=90, include_bias=False)),
    ('std_scaler', StandardScaler())
])

train_input_poly_scaled = poly_scaler.fit_transform(train_input)
val_input_poly_scaled = poly_scaler.transform(val_input)

sgd = SGDRegressor(max_iter = 1, tol = 0, warm_start=True, # fit() 호출될 때 처음부터 시작하지 않고 이전 파라미터에서 훈련 이어감
                  penalty = None, learning_rate = 'constant', eta0=0.0005)

minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000) :
    sgd.fit(train_input_poly_scaled, train_target)
    val_target_predict = sgd.predict(val_input_poly_scaled)
    val_error = mean_squared_error(val_target, val_target_predict)
    if val_error < minimum_val_error :
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = deepcopy(sgd)    
```

# 연습문제

```python
# 12) 조기종료를 사용한 배치 경사 하강법으로 소프트맥스 회귀 구현 (사이킷런 사용 X)
X = iris['data'][:, (2, 3)]
y = iris['target']

X_with_bias = np.c_[np.ones([len(X), 1]), X]
```

```python
test_ratio = 0.2
validation_ratio = 0.2
total_size = len(X_with_bias) # 150

test_size = int(total_size * test_ratio)  # 30
validation_size = int(total_size * validation_ratio) # 30
train_size = total_size - test_size - validation_size # 90

rnd_indices = np.random.permutation(total_size)
```

```python
train_input = X_with_bias[rnd_indices[:train_size]] # (90, 3)
train_target = y[rnd_indices[:train_size]]
val_input = X_with_bias[rnd_indices[train_size : -test_size]]
val_target = y[rnd_indices[train_size : -test_size]]
test_input = X_with_bias[rnd_indices[-test_size:]]
test_target = y[rnd_indices[-test_size:]]
```

```python
def to_one_hot(y):
    n_classes = y.max() + 1
    m = len(y)
    Y_one_hot = np.zeros((m, n_classes))
    Y_one_hot[np.arange(m), y] = 1
    return Y_one_hot
```

```python
train_one_hot = to_one_hot(train_target)
val_one_hot = to_one_hot(val_target)
test_one_hot = to_one_hot(test_target)
```

```python
def softmax(logits):
    exps = np.exp(logits)
    exp_sums = np.sum(exps, axis=1, keepdims=True)
    return exps / exp_sums
```

```python
n_inputs = train_input.shape[1] # 3
n_outputs = len(np.unique(train_target)) # 3
```

```python
eta = 0.01
n_iterations = 5001
m = len(train_input)
epsilon = 1e-7

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = train_input.dot(Theta)
    y_proba = softmax(logits)
    if iteration % 500 == 0:
        loss = -np.mean(np.sum(train_one_hot * np.log(y_proba + epsilon), axis=1))
        print(iteration, loss)
    error = y_proba - train_one_hot
    gradients = 1/m * train_input.T.dot(error)
    Theta = Theta - eta * gradients
```

```python
Theta
```

```python
logits = val_input.dot(Theta)
y_proba = softmax(logits)
y_predict = np.argmax(y_proba, axis=1)

accuracy_score = np.mean(y_predict == val_target)
accuracy_score
```

```python
eta = 0.1
n_iterations = 5001
m = len(train_input)
epsilon = 1e-7
alpha = 0.1  # regularization hyperparameter

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = train_input.dot(Theta)
    y_proba = softmax(logits)
    if iteration % 500 == 0:
        xentropy_loss = -np.mean(np.sum(train_one_hot * np.log(y_proba + epsilon), axis=1))
        l2_loss = 1/2 * np.sum(np.square(Theta[1:]))
        loss = xentropy_loss + alpha * l2_loss
        print(iteration, loss)
    error = y_proba - train_one_hot
    gradients = 1/m * train_input.T.dot(error) + np.r_[np.zeros([1, n_outputs]), alpha * Theta[1:]]
    Theta = Theta - eta * gradients
```

```python
logits = val_input.dot(Theta)
y_proba = softmax(logits)
y_predict = np.argmax(y_proba, axis=1)

accuracy_score = np.mean(y_predict == val_target)
accuracy_score
```

```python
eta = 0.1 
n_iterations = 5001
m = len(train_input)
epsilon = 1e-7
alpha = 0.1  # regularization hyperparameter
best_loss = np.infty

Theta = np.random.randn(n_inputs, n_outputs)

for iteration in range(n_iterations):
    logits = train_input.dot(Theta)
    y_proba = softmax(logits)
    error = y_proba - train_one_hot
    gradients = 1/m * train_input.T.dot(error) + np.r_[np.zeros([1, n_outputs]), alpha * Theta[1:]]
    Theta = Theta - eta * gradients

    logits = val_input.dot(Theta)
    y_proba = softmax(logits)
    xentropy_loss = -np.mean(np.sum(val_one_hot * np.log(y_proba + epsilon), axis=1))
    l2_loss = 1/2 * np.sum(np.square(Theta[1:]))
    loss = xentropy_loss + alpha * l2_loss
    if iteration % 500 == 0:
        print(iteration, loss)
    if loss < best_loss:
        best_loss = loss
    else:
        print(iteration - 1, best_loss)
        print(iteration, loss, "early stopping!")
        break
```

```python
y_proba = softmax(logits)
y_predict = np.argmax(y_proba, axis=1)

accuracy_score = np.mean(y_predict == val_target)
accuracy_score
```
